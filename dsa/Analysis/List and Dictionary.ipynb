## Dictionary

```python
import timeit
import random

for i in range(10000,1000001,20000):
    t = timeit.Timer("random.randrange(%d) in x"%i,
                     "from __main__ import random,x")
    x = list(range(i))
    lst_time = t.timeit(number=1000)
    x = {j:None for j in range(i)}
    d_time = t.timeit(number=1000)
    print("%d,%10.3f,%10.3f" % (i, lst_time, d_time))
```

## Programming Exercises

1. Devise an experiment to verify that the list index operator is O(1)

```python
times = []
for i in range(1000):
    t = timeit.Timer("x[%d]" %i, 
                     "from __main__ import x")
    x = list(range(1000))
    time = t.timeit(number=10000)
    times.append(time)
```

```python
plt.ylim((0, 0.001))
plt.plot(times, '.')
```

2. Devise an experiment to verify that get item and set item are O(1) for dictionaries.

```python
get_times = []
set_times = []
for i in range(1000):
    t = timeit.Timer("d[%d]=%d" %(i, i), 
                     "from __main__ import d")
    t2 = timeit.Timer("d[%d]" %i, 
                     "from __main__ import d")
    d = {x[i]: 0 for i in range(1000)}
    get_time = t.timeit(number=1000)
    get_times.append(get_time)
    set_time = t2.timeit(number=1000)
    set_times.append(set_time)
```

```python
plt.ylim((0, 0.0001))
plt.plot(get_times, '.')
plt.plot(set_times, '.')
```
